```python
import tensorflow as tf 
from tensorflow.keras.models import load_model
import pickle
import pandas as pd 
import numpy as np 
```

```python
## Load the pickle files (trained model , scaler/ onehotencoder/labelencoder)  and do the prediction 

model = load_model('model.h5')

with open('label_encoder_new_vs_repeat.pkl','rb') as file :
    label_encoder_new_vs_repeat=pickle.load(file)
    
with open('onehot_encoder_loantype.pkl','rb') as file :
    onehot_encoder_loantype=pickle.load(file)
    
with open('scaler.pkl','rb') as file :
    scaler=pickle.load(file)
```

```python
input_data = pd.read_csv('Test.csv')
```

```python
input_data = input_data.drop(['ID', 'customer_id', 'country_id', 'tbl_loan_id', 'lender_id'],axis=1)
input_data.head()
```

```python
input_data['loan_type'].unique()
```

```python
# Trouver les lignes où 'loan_type_Type_1' est null
lignes_null = input_data[input_data['loan_type'].isna()]

print("Lignes avec 'loan_type' null :")
print(lignes_null)
```

```python
colonnes_avec_nan = input_data.columns[input_data.isna().any()].tolist()

print("Colonnes avec des valeurs NaN :")
print(colonnes_avec_nan)
```

```python
unknown_categories = ['Type_8', 'Type_3']

# input_data = input_data[~input_data['loan_type'].isin(unknown_categories)]

# Calcul du mode de la colonne 'loan_type'
mode_value = input_data['loan_type'].mode()[0]

# Remplacer les lignes où 'loan_type' est dans unknown_categories par le mode
input_data['loan_type'] = input_data['loan_type'].apply(lambda x: mode_value if x in unknown_categories else x)

```

```python
### loantype column

loantype_encoder = onehot_encoder_loantype.transform(input_data[['loan_type']])
loantype_encoded_df = pd.DataFrame(loantype_encoder.toarray(),columns=onehot_encoder_loantype.get_feature_names_out(['loan_type']))
loantype_encoded_df.head()
```

```python
loantype_encoded_df.isna().sum()
```

```python
input_data = pd.concat([input_data.drop('loan_type',axis=1),loantype_encoded_df],axis=1)
input_data.head()
```

```python
input_data.isna().sum()
```

```python
## Check missing values
# input_data['New_versus_Repeat'].isnull().sum()
input_data['New_versus_Repeat']
```

```python
# mode_value=input_data[input_data['New_versus_Repeat'].notna()]['New_versus_Repeat'].mode()[0]
```

```python
# mode_value
```

```python
# input_data['New_versus_Repeat']=input_data['New_versus_Repeat'].fillna(mode_value)
```

```python
input_data['New_versus_Repeat'] = label_encoder_new_vs_repeat.transform(input_data['New_versus_Repeat'])
input_data.head()
```

```python
input_data['Date_disbursement']=input_data['disbursement_date'].str.split('-').str[2]
input_data['Month_disbursement']=input_data['disbursement_date'].str.split('-').str[1]
input_data['Year_disbursement']=input_data['disbursement_date'].str.split('-').str[0]
```

```python
input_data['Date_due']=input_data['due_date'].str.split('-').str[2]
input_data['Month_due']=input_data['due_date'].str.split('-').str[1]
input_data['Year_due']=input_data['due_date'].str.split('-').str[0]

input_data.head()
```

```python
input_data = input_data.drop(['disbursement_date','due_date'],axis=1)
input_data.head()
```

```python
## change the type from object to int 

input_data['Date_disbursement']=input_data['Date_disbursement'].astype(int)
input_data['Month_disbursement']=input_data['Month_disbursement'].astype(int)
input_data['Year_disbursement']=input_data['Year_disbursement'].astype(int)

input_data['Date_due']=input_data['Date_due'].astype(int)
input_data['Month_due']=input_data['Month_due'].astype(int)
input_data['Year_due']=input_data['Year_due'].astype(int)



```

```python
# Identifier les colonnes avec des valeurs NaN
colonnes_avec_nan = input_data.columns[input_data.isna().any()].tolist()

print("Colonnes avec des valeurs NaN :")
print(colonnes_avec_nan)
```

```python
input_data.head()
```

```python
input_data_scaled = scaler.transform(input_data)
input_data_scaled
```

```python
prediction = model.predict(input_data_scaled)


# Créer un DataFrame avec la colonne 'target'
df_prediction = pd.DataFrame(prediction, columns=['target'])
```

```python
df_prediction.head()
```

```python
# Remplacer les valeurs selon la condition
df_prediction['target'] = df_prediction['target'].apply(lambda x: 0 if x < 0.5 else 1)
df_prediction.head()
```

```python
data_id= pd.read_csv('Test.csv')
data_id.head()
```

```python
data_id = data_id['ID']
data_id.head()
```

```python
data_submission = pd.concat([data_id,df_prediction],axis=1)
data_submission.head()
```

```python
# Sauvegarder le DataFrame dans un fichier Excel
data_submission.to_csv('submission.csv', index=False)

print("Le fichier Excel a été sauvegardé.")
```
